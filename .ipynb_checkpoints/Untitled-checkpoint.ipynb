```python
from tsetmc_webservice import WebserviceClient, Flow
import pandas as pd
from datetime import datetime
import time
```

```python
client = WebserviceClient('fnst.ut.ac.ir', 'fnst.ut.ac')
```

```python
all_trades = []
all_best_limits = []
all_client_types = []
while len(all_trades) < 200:
    print(len(all_trades))
    all_trades.extend([{'time': datetime.now(), **x} for x in client.trade_last_day(Flow.BOURSE)])
    print('.')
    all_best_limits.extend([{'time': datetime.now(), **x} for x in client.best_limits_all_ins(Flow.BOURSE)])
    print('..')
    all_client_types.extend([{'time': datetime.now(), **x} for x in client.client_type()])
    print('...')
    print('----------------------------------')
```
